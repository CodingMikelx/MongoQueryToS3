```python
included_keys = ['user_id', 'profile_id', 'upgrade_id', 'star', 'current_theme', 'time_updated', 'time_created']
result_list = []

for item in data:
    result_item = {k: v for k, v in item.items() if k in included_keys}
    
    # Handle nested 'data' dictionary
    if 'data' in item:
        data_item = item['data']
        
        # Check if 'data_item' is not empty before assigning 'data_result'
        if data_item:
            data_result = {k: v for k, v in data_item.items() if k in included_keys and k not in ['user_id', 'profile_id']}
            
            result_item['data'] = data_result
    
    result_list.append(result_item)
```
